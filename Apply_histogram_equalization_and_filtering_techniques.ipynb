#Apply histogram equalization and filtering techniques.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('/content/1073.jpg', 0)  # Read the image in grayscale

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Apply filtering techniques (e.g., median filtering)
filtered_image = cv2.medianBlur(equalized_image, 5)  # 5x5 median filter

# Display the original, equalized, and filtered images
plt.figure(figsize=(10, 7))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')

plt.subplot(1, 3, 3)
plt.imshow(filtered_image, cmap='gray')
plt.title('Filtered Image')

plt.show()
```

```python
import cv2
import numpy as np

# Load the image
image_path = '/content/14.jpg'
image = cv2.imread(image_path)

# Edge detection using the Canny edge detector
edges = cv2.Canny(image, 100, 200)

# Smoothing (blurring) using a Gaussian filter
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Sharpening
# Create a kernel for sharpening
sharpening_kernel = np.array([[-1, -1, -1],
                              [-1, 9, -1],
                              [-1, -1, -1]])
sharpened = cv2.filter2D(image, -1, sharpening_kernel)

# Save the processed images
output_base_path = '/content/14.jpg'
cv2.imwrite(output_base_path + 'edges.jpg', edges)
cv2.imwrite(output_base_path + 'blurred.jpg', blurred)
cv2.imwrite(output_base_path + 'sharpened.jpg', sharpened)

# Return the paths to the saved images
output_paths = {
    'original': image_path,
    'edges': output_base_path + 'edges.jpg',
    'blurred': output_base_path + 'blurred.jpg',
    'sharpened': output_base_path + 'sharpened.jpg'
}

output_paths

import matplotlib.pyplot as plt

# Load the processed images
edges_image = cv2.imread(output_paths['edges'])
blurred_image = cv2.imread(output_paths['blurred'])
sharpened_image = cv2.imread(output_paths['sharpened'])

# Convert BGR to RGB for plotting
edges_image_rgb = cv2.cvtColor(edges_image, cv2.COLOR_BGR2RGB)
blurred_image_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)
sharpened_image_rgb = cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB)

# Plot the images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# Edge Detection
axes[0].imshow(edges_image_rgb)
axes[0].set_title('Edge Detection')
axes[0].axis('off')
# Blurred Image
axes[1].imshow(blurred_image_rgb)
axes[1].set_title('Blurred Image')
axes[1].axis('off')
# Sharpened Image
axes[2].imshow(sharpened_image_rgb)
axes[2].set_title('Sharpened Image')
axes[2].axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```
